```python
# Data Understanding (Read CSV, Get X and y)

# Importing numpy with an alias np
import numpy as np

# Importing pyplot from matplotlib with an alias plt
import matplotlib.pyplot as plt

# Importing pandas with an alias pd
import pandas as pd


# Importing data as datafile
datafile = pd.read_csv('diagnose2.csv')
```

### 3. Data preparation
<p style='text-align: justify;'> Data preparation is the method of cleaning and transforming the raw data before processing and analysing so as to data experts can run machine learning algorithms through it to discover insights or make predictions. This phase is also known as data preprocessing.
</p>

#### 3.1 Selecting the features and target
<p style='text-align: justify;'> For the given data an id, gender or location may not be helpful to make a diagnosis prediction or at least for this analysis. So, the first three fields will be excluded from analysis. The fields from column 3 to 16 are the main features of this dataset which will be used to predict the future y or diagnosis. Thus, the following codes will perform this task.</p>

```python
# Specify the features X and the target y

# Features are from column 3 to 15 because 16 is not inclued.
# X is taken capital as it is representing the all the x's from x1 to x13
X = datafile.iloc[:, 3:16].values

# The target or future for the given dataset is diagosis or column 16.
y = datafile.iloc[:, 16].values

# Displaying the features
print(X, "\n")
# Displaying the future or target
print(y)

#Displaying the statistical report
display(datafile.describe())
```

#### 3.2 Handling missig values
<p style='text-align: justify;'> Handling the missing values is a very important part of data preparation. Analytics Vidhya argues that a majority of machine learning algorithms are unsuccessful when dataset has missing values. It also claims that the output may result in creating biased machine learning model becaues of the missing values that will cause inaccurate results once these are not properly handled.</p>

<p style='text-align: justify;'> The given dataset has some missing values and these needs to be dealt with. The machine learning tool SimpleImputer tool will be used to handle these data. It will replace all the cells containing missing values with mean of their respective columns. Besides, there are several ways to do this like the missing values can also be replaced with median, mode.</p>

```python
# Importing SimpleImputer from sklearn.impute
from sklearn.impute import SimpleImputer

# Creating a SimpleImputer object and specifying how missing values are represented
# Strategy ='Median' or 'Mode' can also be used.
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fitting this to the datase by specifying the relevant features like the columns
X[:,0:16] = imputer.fit_transform(X[:,0:16])

# Printing X to check if the above operations worked.
print(X)
```

#### 3.3 Feature Scaling
<p style='text-align: justify;'> Feature scaling is a good process for better results. It is a process for normalising the variety of independent variables or features in data. According to DIM, scaling of the data helps machine learnig models to understand and learn problems easily. From sklearn.preprocessing, StandardScaler will be used to perform this operation.</p>

```python
# Feature Scaling
# Importing StandardScaler from sklearn.preprocessing
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X[:,3:16] = sc.fit_transform(X[:,3:16])
print(X)
```

#### 3.4 Spliting Data
<p style='text-align: justify;'>
The dataset will be splitted into training set and test set.<br> I. The proportion of data taken to train the model is called training set. The model is expected to observe and learn from the training set by optimizing any of its parameters. <br><br>II. The portion of data used to test the final model is called the testing set, and it is compared against the earlier sets of data. The testing set acts as an evaluation of the final model and algorithm. <br><br>The percentage of the train and test set may vary. But for the given dataset, 80% data will be used for training set and 20% data for test set. </p>

```python
# Importing train_test_split from sklearn.model_selection
from sklearn.model_selection import train_test_split

# Splitting data in training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state =0 )

#Checking if the data seperation is successfully done.
print(X_train)
print(y_train)

print(X_test)
print(y_test)
```

#### 3.5 Feature Selection
<p style='text-align: justify;'> Simplilearn defines feature selection as the process of decreasing the input variable to a model by considering only compatible data and avoiding noise in data.The goal of feature selection in machine learning (ML) is to find the best set of characteristics that allow engineers to create constructive and usable models of the subject they are attempting to research.
</p>

```python
# Feature Selection
# Importing SelectKBest from sklearn.feature_selection
from sklearn.feature_selection import SelectKBest

# Importing mutual_info_regression from sklearn.feature_selection
from sklearn.feature_selection import mutual_info_regression

# Identifying the relevant features based on the training set
select = SelectKBest(mutual_info_regression, k = 'all').fit(X_train, y_train)

# Transforming the training and test inputs
X_train = select.transform(X_train)
X_test = select.transform(X_test)

# Cheking the success of the task done above here
print (X_train)
```

#### 3.6 K-fold cross-validation
<p style='text-align: justify;'>Cross-validation is a technique for evaluating a machine learning model and testing its performance. It is commonly used in applied ML tasks. It helps to compare and select an appropriate model for the specific predictive modelling problem. Here K denotes the number of groups which the data sample is divided into.<br>  
Here we are taking the value of k as 10 which means there will be 10 number of groups. It can be called as 10-fold cross-validation.
<p/>

```python
# Importing the required modules
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold, cross_validate
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
# Creating a Naive Bayes Classifier
nb = GaussianNB()
# Applying k-Fold Cross Validation
cv = KFold(n_splits=10, shuffle=True, random_state = 0)
for kfs in range(1,11):
    select = SelectKBest(score_func=f_classif, k=kfs).fit(X_train, y_train)
    
# identify the relevant features based on the training data
# transform the training and test inputs
X_trainFS = select.transform(X_train)
X_testFS = select.transform(X_test)
scores = cross_validate(estimator = nb, X = X_trainFS, y = y_train, cv = cv)
print(scores["test_score"].mean())
```

## 4. Data Modelling
<p style='text-align: justify;'>A model in Machine Learning abstractly depicts a decision process. The fundamental purpose of the concept is to enable decision-making automation, which is frequently used in business. In certain circumstances, the model aids in comprehending the modelled process. Machine Learning models are data-driven mathematical algorithms that are "trained" with data. #### 4.1 Random Forest Classifier
<p style='text-align: justify;'> Random Forest is a classifier that uses many decision trees on different subsets of the input dataset and averages the results to increase the dataset's predicted accuracy.This model's strength is in its ability to construct several trees from the features with various sub-features. Since each tree's features are chosen at random, the trees do not grow very deep and just concentrate on the set of features.</p>

```python
# Learn Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
rf.fit(X_train, y_train)
# Predict test cases
y_pred = rf.predict(X_test)
# Construct the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

#### 4.1.1 Precision Calculation for Random Forest

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('\nAccuracy: {:.2f}\n'.format(accuracy_score(y_test, y_pred)))

print('Micro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='micro')))
print('Micro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='micro')))
print('Micro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='micro')))

print('Macro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='macro')))
print('Macro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='macro')))
print('Macro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='macro')))

print('Weighted Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='weighted')))
print('Weighted Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='weighted')))
print('Weighted F1-score: {:.2f}'.format(f1_score(y_test, y_pred, average='weighted')))
```

#### 4.2 K-Nearest Neighbors (K-NN)
<p style='text-align: justify;'> IBM describes K-Nearest Neighbours, sometimes referred to as KNN or k-NN, as a supervised learning classifier that employs proximity to produce classifications or predictions about the grouping of a single data point. Although it can be applied to classification or regression issues, it is commonly employed as a classification algorithm because it relies on the idea that comparable points can be discovered close to one another.</p>

```python
# Learn k-NN Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
# Predict test cases
y_pred = knn.predict(X_test)
# Construct the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

#### 4.2.1 Precision Calculation for K-NN Classifier

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('\nAccuracy: {:.2f}\n'.format(accuracy_score(y_test, y_pred)))

print('Micro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='micro')))
print('Micro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='micro')))
print('Micro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='micro')))

print('Macro Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='macro')))
print('Macro Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='macro')))
print('Macro F1-score: {:.2f}\n'.format(f1_score(y_test, y_pred, average='macro')))

print('Weighted Precision: {:.2f}'.format(precision_score(y_test, y_pred, average='weighted')))
print('Weighted Recall: {:.2f}'.format(recall_score(y_test, y_pred, average='weighted')))
print('Weighted F1-score: {:.2f}'.format(f1_score(y_test, y_pred, average='weighted')))
```
